# Campo Gravitacional

```python
xmax = 200 #Límites del Eje X
xmin = -xmax

NX = 80 #Divisiones en X
NY = 80 #Divisiones en Y

ymax = 200 #Límites del Eje Y
ymin = -ymax

x1 = linspace(60, xmax, NX) #Creación de Linspaces
y1 = linspace(60, ymax, NY)

x2 = linspace(xmin, -60, NX)
y2 = linspace(ymin, -60, NY)

x = x2 + x1
y = y2 + y1

X, Y = meshgrid(x, y) #Se crea la cuadrícula que llena el gráfico.

Cx = -XX1 / (XX1**2 + YY1**2)**(3/2)
Cy = -YY1 / (XX1**2 + YY1**2)**(3/2)

gridExternoX = []
gridExternoY = []
gridInternoX = []
gridInternoY = []
for i in range(80):
    arregloActualX = X[i]
    arregloActualY = Y[i]
    for j in range(len(X[0])):
        if(arregloActualX[j]**2 + arregloActualY[j]**2 >= 25):
            gridExternoX.append(arregloActualX[j])
            gridExternoY.append(arregloActualY[j])
        else:
            gridInternoX.append(arregloActualX[j])
            gridInternoY.append(arregloActualY[j])
            
arrayExternoX = asarray(gridExternoX)
arrayExternoY = asarray(gridExternoY)
CExternoX = -1*(arrayExternoX) / (arrayExternoX**2 + arrayExternoY**2)**(3/2) 
CExternoY = -1*(arrayExternoY) / (arrayExternoX**2 + arrayExternoY**2)**(3/2)
print(arrayExternoX)
```

# Campo Gravitacional dentro de la Tierra

```python
p = linspace(-5, 5, 3) #Crear una cuadrícula en este rango.
XX = []
YY = []
for i in range(len(p)):
    for j in range(len(p)):
        XX.append(p[i])
        YY.append(p[j])
        
XX1 = asarray(XX) #Convertir la lista en arreglo umérico.
YY1 = asarray(YY)
CXX1 = -100*XX1 #Dar las coordenadas X y Y de cada elemento.
CYY1 = -100*YY1
figure(figsize(5,5))
scatter(XX,YY) #Scatter para visualizar
```

```python
#Construcción de la Tierra
t = linspace(0,2*pi, 100) #Linspace para la tierra.
figure(figsize=(5,5))
plot(7*cos(t), 7*sin(t), color = "Blue", lw = 2, label = "Superficie Terrestre") #Círculo terrestre.
QP = quiver(XX1,YY1,CXX1,CYY1) #Campo dentro de la tierra.
legend(bbox_to_anchor=(1.6,1))
```

```python
pI1 = linspace(-50, -16, 10) #Cuadrícula Izquierda
pI2 = linspace(-50,12,20) 
XXI = []
YYI = []
for i in range(20):
    for j in range(10):
        XXI.append(pI1[j])
        YYI.append(pI2[i])
aXXI = asarray(XXI)
aYYI = asarray(YYI)

pD1 = linspace(16,50,10) #Cuadrícula Derecha
pD2 = linspace(-50, 13, 20)
XXD = []
YYD = []
for i in range(20):
    for j in range(10):
        XXD.append(pD1[j])
        YYD.append(pD2[i])
aXXD = asarray(XXD)
aYYD = asarray(YYD)

pAr1 = linspace(-50, 50, 30) #Cuadrícula Arriba
pAr2 = linspace(15, 50, 16)
XXAr = []
YYAr = []
for i in range(30):
    for j in range(16):
        XXAr.append(pAr1[i])
        YYAr.append(pAr2[j])
aXXAr = asarray(XXAr)
aYYAr = asarray(YYAr)

pAb1 = linspace(-13,13,8)
pAb2 = linspace(-15,-50,16)
XXAb = []
YYAb = []
for i in range(8):
    for j in range(16):
        XXAb.append(pAb1[i])
        YYAb.append(pAb2[j])
aXXAb = asarray(XXAb)
aYYAb = asarray(YYAb)
scatter(XXI,YYI)
scatter(XX1,YY1, color = "Red")
scatter(XXD,YYD, color = "Blue")
scatter(XXAr, YYAr)
scatter(aXXAb, aYYAb)
```
