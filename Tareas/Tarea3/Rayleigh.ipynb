###Introducción
En este experimento se buscó poder, primero, obtener datos al aprender a compilar ejecutables de C correctamente, y segundo, utilizar dichos datos obtenidos para realizar un análisis con las funciones básicas de Python y el uso de la librería MatPlotLib aprendidas en clase. Entre dichas funciones utilizadas se encuentran **hist()**, para realizar histogramas de frecuencia, y **curve_fit**, para realizar curvas de ajuste más óptimo dada una función general y datos para ajustarse. Estas funciones, con sus diferentes opciones y configuraciones, se aplicaron para comprobar experimentalmente la **Distribución de Rayleigh**. La hipótesis sobre la cual se trabajará es que, bajo dicho potencial, y en estado de equilibrio, las velocidades de todas las partículas se comportatrá según dicta la distribución de Rayleigh.

Este experimento toma como base teórica la **Distribución de Rayleigh**. Dentro del ámbito de la probabilidad, esta distribución representa una densidad de probabilidad con parámetro de escala $\sigma$. Este parámetro representa  que tan dispersa es la distribución. Siendo entonces f(x,$\sigma$) una función de densidad de probabilidad, al realizar la integral de esta en un intervalo dado, se obtiene la probabilidad de que el suceso ocurra obteniendo un valor en ese mismo intervalo. La función de densidad de probabilidad de Rayleigh viene definida por:
$$f(x,\sigma )= \frac{x}{\sigma^2} e^{\frac{-x^2}{2\sigma^2}}$$
La gráfica de dicha función se muestra a continuación, ejemplo tomado de Wikipedia:

Para este caso, se busca comprobar la teoría que dicta que las partículas de un gas bajo el efecto del potencial de **Lennard-Jones**, que dicta el potencial total bajo le efecto de la fuerza atractiva de Van Der Waals a grandes distancias y la fuerza repulsiva de soperposición de orbitales a cortas distancias, tendrán velocidades determinadas por la distribución de Rayleigh. En la simulación de partículas de un gas realizada con los archivos ejecutables utilizados, con características especiales de stepAvg = 2, stepLimit = 500 y initUcell = 40 40 como condiciones de entrada, este potencial será el que dicte las posiciones y velocidades de las partículas involucradas durante cada instante de tiempo. ###Obtención y Análisis de Datos
####Velocidades de las Partículas
Para comenzar con este experimento, se descargó el software de simulación indicado en la guía, a la vez que se realizaron las modificaciones indicadas anteriormente. Al compilar y ejecutar los archivos de simulación indicados por la guía, se creó un archivo denominado **Rayleigh.csv** que contiene posiciones en X y Y, a la vez que velocidades en X y Y de cada una de las 1600 partículas simuladas en cada instante. Una vez se tiene este archivo, se importó a este notebook para ser utilizado. Para empezar el análisis, se separaran entonces las distintas columnas en arrays distintos, para poder realizar análisis por separado. Esto se realizó como se muesra a continuación.

````python
%pylab inline
from matplotlib import animation
from scipy.optimize import curve_fit
import time
from IPython.display import display, clear_output
````

````python
#Se importa al notebook el archivo .csv
datos=genfromtxt("Rayleigh.csv",delimiter=",")
````

````python
#Se separan en arrays distintos las posiciones y velocidades en cada eje.
x=datos[:,0]
y=datos[:,1]
vx=datos[:,2]
vy=datos[:,3]
````

Para continuar, se busca poder realizar un histograma de frecuencia de los disntios valores de rapidez de las partículas en el equilibrio. Por este motivo, se debe analizar el último instante de la simulación y tomar la rapidez de las 1600 partículas en este instante. Dado que los arrays construidos anteriormente contienen los datos correspondientes a componentes de la velocidad, se calcula un nuevo arreglo de valores que contenga el módulo de la velocidad de cada partícula. Por practicidad, dicho cálculo se implementa directamente dentro del comando **hist()** de manera que con los valores contenidos en este se realice un histograma normalizado de la rapidez de todas las partículas simuladas en el último instante.

````python
#Se declara la figura que contiene al histograma.
figure(figsize=(10,10))
#Se construye un histograma normalizado de 15 columnas de la rapidez de las partículas en el equilibrio.
hist((vx[-1601:-1]**2+vy[-1601:-1]**2)**(0.5), bins = 50, normed = True, label = "Frecuencia Normalizada de Rapidez")
#Se construye la estética del histograma, como ejes y leyenda.
title("Frecuencia de Rapideces")
ylabel("Frecuencia Normalizada")
xlabel("Rapidez")
legend()
show()
````

Una vez se visualiza claramente el histograma de frecuencia, es posible ver que, a grandes rasgos, el comportamiento de las rapideces se asemeja a la distribución de **Rayleigh** mostrada anteriormente. Sin embargo, para poder confirmar esta observación, se procede entonces a realizar un ajuste de curva a los datos presentados por el histograma. Y debido a la hipótesis de este trabajo, la función que se desea ajustar es la que define la distribución de **Rayleigh**, mostrada en el marco teórico.

````python
#Realización del Ajuste a la Distribución de Rayleigh.
#Se crea un arreglo que contenga los datos presentados por el histograma.
variable = hist((vx[-1601:-1]**2+vy[-1601:-1]**2)**(0.5), bins = 50, normed = True)
#Se declaran dos arreglos que contengan los datos del eje X y el eje Y.
datosY = variable[0]
datosX = delete(variable[1], -1)
#Se construye la función que se desea ajustar, con parámetro de escala sigma cuadrado a estimarse.
def ajuste(x,a):
    return (x/a)*(exp((-x**2)/(2*a)))
fitpars,covmat= curve_fit(ajuste,datosX,datosY)
````

Habiendo realizado el ajuste con la función **curve_fit** de los datos mostrados por el histograma a la forma general de la distribución de **Rayleigh**, se procede entonces a mostrarlo gráficamente junto con el histograma para poder determinar que tan acertado y óptimo resulta el ajuste, para poder concluir acerca de la hipótesis inicial.

````python
#Se grafican el histograma y el ajuste realizado.
#Se crea un arreglo como variable independiente.
t = linspace(0,4,100)
#Se crea la figura que continene las gráficas.
figure(figsize=(10,10))
#Se grafica el histograma de nuevo.
hist((vx[-1601:-1]**2+vy[-1601:-1]**2)**(0.5), bins = 50, normed = True, label = "Frecuencia Normalizada de Rapidez")
#Se grafica la curva de ajuste obtenida.
plot(t, ajuste(t, *fitpars), color = "red", linewidth = 2, label = "Distribución de Ryleigh con $\sigma^2$ = 0.9906")
plot(t, ajuste(t, 1), color = "Black", linewidth = 2, label = "Distribución de Ryleigh con $\sigma^2$ = 1")
#Se construye la estética de la gráfica.
title("Densidad de probabilidad")
ylabel("Probabilidad")
xlabel("Rapidez")
legend()
````

De la gráfica anterior se puede observar que, a criterio de observación, los datos de frecuencia de rapideces mostrados por el histograma se ajustan de manera bastante a aproximada a la función de densidad de probabilidad de **Rayleigh** con parámetro de escala igual a 0.96798. Por lo anterior, se podría llegar a afirmar que la hipótesis inicial es correcta, y las rapideces de las partículas en el instante de equilibrio de la simulación se comportan como la densidad de probabilidad de **Rayleigh** mostrada a continuación:
$$f(x,)= \frac{x}{0.9906} e^{\frac{-x^2}{1.9812}}$$

En la gráfica anterior también se muestra la curva del modelo teórico, donde se sabe que el parámetro $\sigma$ es 1. Es posible entonces ver gráficamente la similitud existente entre el modelo teórico y el resultado del ajuste, presentando un error del 0.94$\%$.

````python
#Se grafica la dispersión de las posiciones del último instante.
scatter(x[-1601:-1],y[-1601:-1])
xlim(-40,40)
ylim(-40,40)
title("Posiciones de las Partículas en el Último Instante")
xlabel("Posición en X")
ylabel("Posición en Y")
````

Como es posible ver en la figura anterior, en el último instante las partículas se han dispersado de manera más notable a como se había visto hasta el instante 30 de la simulación. En este instante de tiempo, es posible ver gran cantidad de franjas vacias en el area de simulación, especialmente en el centro y alrededor del origen. Por esta misma razón es posible ver que hacia los extremos hay una mayor presencia de partículas. Para entrar a analizar esto mejor, se realiza un histograma de las magnitudes de distancia de cada partícula al origen durante el último instante.

````python
#Histograma de Posiciones
hist((x[-1601:-1]**2 + y[-1601:-1]**2)**(0.5),bins = 30, normed = True, range =(0,35))
title("Frecuencia de Posiciones")
xlabel("Magnitud de Distancia")
ylabel("Densidad de probabilidad")
show()
````

Como se había observado inicialmente, a medida que la distancia radial desde el origen del gráfico aumenta, la probabilidad de hallar partículas aumenta, específicamente linealmente. Esto corresponde teóricamente a la variación del área a medida que aumenta el radio para un anillo, el cual es $2\pi r dr$, por lo que se procede ahora a determinar el ajuste lineal al comportamiento de estos datos.

````python
#Realización del Ajuste a la Distribución de Rayleigh.
#Se crea un arreglo que contenga los datos presentados por el histograma.
variable1 = hist((x[-1601:-1]**2 + y[-1601:-1]**2)**(0.5),bins = 30, normed = True, range =(0,35))
#Se declaran dos arreglos que contengan los datos del eje X y el eje Y.
posY = variable1[0]
posX = delete(variable1[1], -1)
#Se construye la función que se desea ajustar, con parámetro de escala sigma cuadrado a estimarse.
def ajuste2(x,a,b):
    return (a*x + b)
fitpars,covmat= curve_fit(ajuste2,posX,posY)
````

````python
#Se grafican el histograma y el ajuste realizado.
#Se crea un arreglo como variable independiente.
u = linspace(0,40,100)
#Se crea la figura que continene las gráficas.
figure(figsize=(10,10))
#Se grafica el histograma de nuevo.
hist((x[-1601:-1]**2 + y[-1601:-1]**2)**(0.5),bins = 30, normed = True, label = "Frecuencia Normalizada de Posiciones", range = (0,35))
#Se grafica la curva de ajuste obtenida.
plot(u, ajuste2(u, *fitpars), color = "red", linewidth = 2, label = "y=0.00163x + 0.000967")
#Se construye la estética de la gráfica.
title("Densidad de probabilidad")
ylabel("Probabilidad")
xlabel("Posición")
xlim(0,35)
legend(bbox_to_anchor=(0.52,1))
````

Como es posible ver en la figura anterior, el ajuste de una recta con parámetros mostrados en la gráfica representa de manera más apropiada el comportamiento de los datos de posición. Esto comprueba lo anteriormente sugerido acerca de la distribución de las partículas a lo largo y ancho del area. Dado que se analiza la frecuencia normalizada de la magnitud de distancia del origen a cada una de las partículas de manera radial, un incremento lineal de esta significa que a mayor radio, más partículas se encuentran. También, siendo esta correlación lineal, significa que la variación es constante, por lo que las partículas se encuentran uniformemente distribuidas en el area, ya que no hay cambios súbitos de concentración de partículas a medida que se aumenta la posición R.
