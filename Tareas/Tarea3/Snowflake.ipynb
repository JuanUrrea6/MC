# Dibujar el copo de Koch con Turtle

Para realizar este punto, se utilizó la librería **Turtle** de Python, la cual permite una graficación interactiva basada en comandos relativamente sencillos en términos de dibujo y orientaciones angulares de dichos dibujos. Para obtener la graficación de cualquier paso del copo de nieve de Koch, ejecute la celda a continuación, cambiando el parámetro de **paso** de la función **snowflake()** al final del código.

La base de esta implementación fue dividir el copo en tres fragmentos, cada lado del triángulo inicial, y hacer con tres tortugas el mismo procedimiento, formando 60 grados una con la anterior para construir el triángulo. Para abordar el problema, tomé como guía lo explicado en http://openbookproject.net/thinkcs/python/english3e/recursion.html, donde explican básicamente la recursividad que podría resolver el problema de realizar fractales. También se utilizó la documentación oficial de la librería **Turtle**, que se encuentra en https://docs.python.org/2/library/turtle.html.

````python
turtle.setup(400, 400) #Se crea el tamaño de la ventana de Turtle.
ventana = turtle.Screen() #Se crea la ventana para Turtle.
ventana.bgcolor("white") #Se crea un fondo blanco.
ventana.title("Snowflake") #Se declara el título.

tortuga = turtle.Turtle() #Se crea la tortuga 1.
tortuga.speed(7) #Se declara la velocidad de la tortuga 1.
tortuga.hideturtle()#Se esconde la tortuga 1.
tortuga2 = turtle.Turtle() #Se crea la tortuga 2.
tortuga2.speed(7) #Se declara la velocidad de la tortuga 2.
tortuga2.hideturtle() #Se esconde la tortuga 2.
tortuga3 = turtle.Turtle() #Se crea la tortuga 3.
tortuga3.speed(7) #Se declara la velocidad de la tortuga3.
tortuga3.hideturtle() #Se esconde la tortuga 3.
tortuga.color("darkgreen") #Se asigna el color de la tortuga 1.
tortuga.penup() #Se desplaza la tortuga 1 a la posición deseada sin dejar marca.
tortuga.backward(150)
tortuga.left(90)
tortuga.forward(90)
tortuga.right(90)
tortuga.pendown()

tortuga2.color("blue") #Se asigna el color a la tortuga 2.
tortuga2.penup() #Se desplaza la tortuga 2 a la posición deseada a 150 unidades horizontalmente de la 1.
tortuga2.backward(150)
tortuga2.left(90)
tortuga2.forward(90)
tortuga2.right(90)
tortuga2.pendown()
tortuga2.right(60) #Se desplaza la tortuga 3 a la posición deseada sin dejar marca.

tortuga3.color("red") #Se asigna el color a la tortuga 3.
tortuga3.penup() #Se desplaza la tortuga 3 a la posición deseada sin dejar marca, en el último vértico del triángulo.
tortuga3.forward(150)
tortuga3.left(90)
tortuga3.forward(90)
tortuga3.right(90)
tortuga3.pendown()
tortuga3.right(120) #Se gira la tortuga 3 hacia un lado del triángulo.
size = 20 #Se declara el tamaño de la figura.
def snowflake(t1, t2, t3, paso, largo): #Se crea una función que construya el fractal. Recibe de parámetro el paso y el largo.
    size = 20
    if(paso==0): #Se crea el paso base.
        t1.forward(largo)
        t2.forward(largo)
        t3.forward(largo)
    
    else: #Se crea la recursividad.
        snowflake(t1, t2, t3, paso - 1, largo/3) #Recorrer el enésimo tercio del enésimo-1 tercio del paso anterior.
        t1.left(60) #Las tres tortugas rotan para formar el lado del triángulo.
        t2.right(60)
        t3.left(60)
        size = size*3 #Recalibrar la escala.
        snowflake(t1, t2, t3, paso - 1, largo/3) 
        t1.right(120) #Rotar para formar el primer vértice.
        t2.left(120)
        t3.right(120)
        size = size*3 #Recalibrar la escala.
        snowflake(t1, t2, t3, paso - 1, largo/3) 
        t1.left(60) #Rotar para formar retornar a la recta original.
        t2.right(60)
        t3.left(60)
        size = size*3 #Recalibrara la escala.
        snowflake(t1, t2, t3, paso - 1, largo/3) #Recorrer primer tercio.

snowflake(tortuga, tortuga2, tortuga3, 1, 300) #Crear snowflake de largo 300 con pasos como parámetro.
#En el segundo parámetro, se puede introducir el paso deseado.

ventana.exitonclick() #Se declara la opción de cerrar la ventana.
````
