# Rendija Cuadrada

Para comenzar, se crea una matriz de 1024x1024 toda rellena de ceros. Esta se modificará para cumplir los requisitos del enunciado.

```python
#Crear la matriz de 1024x1024 con ceros.
matriz1 = zeros((1024,1024), dtype = np.float_)
```

```python
for i in range(496,528): #Se convierte el cuadrado central de lado 32 en 0.
    for j in range(496,528):
        matriz1[i,j] = 1
```

A continuación, se toma esta rejilla y se realiza la transofmrada de Fourier en dos dimensiones a la matriz creada. Luego, se aplica la función shift para mover las frecuencias 0 de posición. Por último, se halla el logaritmo del módulo al cuadrado de esta nueva matriz, habiéndole sumado 1 a cada término para evitar problemas al hallar el logaritmo.

```python
matriz2 = fft2(matriz1) #Transformada discreta de Fourier.
```

```python
matriz3 = fftshift(matriz2) #Cambio de coordenadas a Fourier.
```

```python
#Tomar cuadrado de lado 200
cuadrado = zeros((200,200), dtype = np.complex)
for i in range (200):
    for j in range(200):
        cuadrado[i,j] = matriz3[412 + i, 412 + j]
```

```python
#Hallar el logaritmo del módulo al cuadrado de los datos tomados.
logaritmo = zeros((200,200), dtype = np.float)
for i in range(200):
    for j in range(200):
        actual = cuadrado[i,j] + 1 #Evitar problemas al logaritmo.
        logaritmo[i,j] = log(abs(actual)**2)
```

```python
figure(figsize=(10,5))
graf1 = subplot2grid((1,2), (0,1), colspan=1)
imshow(logaritmo) #Se muestran los lados del logaritmo.
title("Patrón de Difracción\n Apertura Cuadrada\n (Log Resaltado)")
graf2 = subplot2grid((1,2), (0,0), colspan=1)
imshow(matriz1, cmap = 'hot')
savefig("AperturaCuadrada.png")
```

# Dos Rendijas Rectangulares

```python
#Crear las dos rendijas en el cuadrado de 1024x1024
#Crear la matriz de 1024x1024 con ceros.
matrizNueva = np.zeros((1024,1024), dtype = np.float_)
```

```python
for i in range(20):
    for j in range(200):
        matrizNueva[412 + j, 460 + i] = 1
        matrizNueva[412 + j, 564 + i] = 1
```

```python
matrizNueva2 = fft2(matrizNueva) #Transformada de fourier discreta a matriz
```

```python
matrizNueva3 = fftshift(matrizNueva2) #cambio de posiciones
```

```python
#Tomar cuadrado de lado 200
cuadrado2 = zeros((128,128), dtype = np.complex)
for i in range (128):
    for j in range(128):
        cuadrado2[i,j] = matrizNueva3[448 + i, 448 + j]
```

```python
moduloNuevo = zeros((128,128), dtype = np.float) #Tomar módulos
for i in range(128):
    for j in range(128):
        moduloNuevo[i,j] = abs(cuadrado2[i,j])**2
```

```python
figure(figsize=(10,5)) #Visualización
graf1 = subplot2grid((1,2), (0,1), colspan=1)
imshow(moduloNuevo, cmap = 'jet')
title("Patrón de Difracción de Doble Rendija")
graf2 = subplot2grid((1,2), (0,0), colspan=1)
imshow(matrizNueva, cmap = 'hot')
savefig("DobleApertura.png")
```

# Rendija Circular

```python
#Crear círculo de radio 10.24
r = linspace(502,522,100)
s = linspace(502,522,100)
circuloX = []
circuloY = []
for i in range(len(r)):
    for j in range(len(s)): #Crear condiciones según ecuación del círculo.
        if((r[i] - 512)**2 + (s[j]-512)**2 <= 100):
            circuloX.append(r[i])
            circuloY.append(s[j])  
```

```python
matrizCirculo = zeros((1024,1024), dtype = float)
for p in range(len(circuloX)):
    matrizCirculo[circuloX[p], circuloY[p]] = 1
```

```python
matrizCirculo2 = fft2(matrizCirculo) #Transformada de fourier a matriz
```

```python
matrizCirculo3 = fftshift(matrizCirculo2) #Cambio de posiciones
```

```python
#Hallar el logaritmo del módulo al cuadrado de los datos tomados.
logaritmoCirculo = zeros((1024,1024), dtype = np.float)
for i in range(1024):
    for j in range(1024):
        actual = matrizCirculo3[i,j] + 1 #Evitar problemas al logaritmo.
        logaritmoCirculo[i,j] = log(abs(actual)**2)
```

# Gráficas y Análisis

```python
filaCentral = abs(matrizCirculo3[512,:])**2/abs(max(matrizCirculo3[512,:]))**2
abscisas = 2*pi*(0.01)*linspace(-512,512,1024) #Tomar fila central y abscisas
```

```python
scatter(abscisas,filaCentral) #Visualizar
xlim(0,10)
ylim(0,1)
xlabel("$2 \pi Rj/ L$")
ylabel("$I/I_0$")
```

# Gráficas en Panel

```python
figure(figsize=(20,15)) #Organizar panel.
subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
scatter(abscisas,filaCentral)
xlim(0,10)
ylim(0,1)
xlabel("$2 \pi Rj/ L$")
ylabel("$I/I_0$")
subplot2grid((3,2),(1,0),rowspan=2)
imshow(matrizCirculo, cmap = 'gray')
subplot2grid((3,2),(1,1),rowspan=2)
imshow(logaritmoCirculo, cmap = 'gray')
title("Patrón de Interferencia Apertura Circular\n Patrón de Airy\n Log Resaltado")
tight_layout()
#savefig("PanelCirculo.png")
```

# Hallar Máximos y Mínimos con Interpolación

Para determinar los máximos y mínimos de los datos entre 0 y 12, usando interpolación, se procedió como se muestra a continuación. Se dividió a trozos el array de datos obtenido anteriormente, usando como criterio la gráfica, dividiendo en intervalos donde se puede apreciar que existe un mínimo o máximo local, de manera que dentro del trozo se volverían máximos y mínimos globales. Una vez se obtienen estos trozos, junto con su trozo correspondiente de abscisas, se utiliza la función **interp1d** para plantear una interpolación de tipo cúbica para modelar cada grupo de datos. Una vez se obtiene esta función, simplemente se utilizan las funciones **max()** y **min()** para obtener estos nuevos máximos y mínimos globales.

```python
scatter(abscisas[512:702], filaCentral[512:702]) #Visualizar los datos para hacer los cortes.
xlim(0,13)
ylim(0,1)
```

```python
corte1 = filaCentral[512:550] #Primer Corte
corte2 = filaCentral[550:600] #Segundo Corte
corte3 = filaCentral[600:640] #Tercer Corte
corte4 = filaCentral[640:680] #Cuarto Corte
corte5 = filaCentral[660:702] #Quinto Corte
```

```python
corte11 = abscisas[512:550] #Se realizan los cortes a las abscisas.
corte22 = abscisas[550:600]
corte33 = abscisas[600:640]
corte44 = abscisas[640:680]
corte55 = abscisas[660:702]
```

```python
scatter(corte11,corte1, color = "Blue", label = "Primer Corte") #Graficar cada set de datos cortados.
scatter(corte22,corte2, color = "Red", label = "Segundo Corte")
scatter(corte33, corte3, color = "Green", label = "Tercer Corte")
scatter(corte44, corte4, color = "Yellow", label = "Cuarto Corte")
scatter(corte55, corte5, color = "Black", label = "Quinto Corte")
title("Función Dividida en Trozos")
xlim(0,12)
ylim(0,1)
legend()
```

En la gráfica anterior es posible ver la descomposición del arreglo inicial en los distintos cortes planteados. Puede que algunos de estos se sobrepongan, sin embargo esto no ocurre en los puntos máximos y mínimos. Ahora, se creará un array de máximos y mínimos para poder realizar la comparación con los valores teóricos dados por el enunciado, que se encuentran en el arreglo de datos teóricos.

```python
datosTeoricos = [1, 0, 0.0175, 0, 0.0042, 0, 0.0016]
```

```python
datosExperimentales = [] #Arreglo de máximos y mínimos
```

## Primer Corte

Para este corte de datos y para todos los demás se utilizará una interpolación cúbica para hallar los máximos y mínimos locales, haciendo uso de la funcion **interp1d()**.

```python
ajuste1 = interp1d(corte11,corte1, kind = 'cubic') #Interpolación cúbica a datos.
```

```python
plot(corte11, ajuste1(corte11), color = "Red", label = "Interpolación\n Cúbica", ls = "-", lw = 1.5)
scatter(corte11,corte1, label ="Datos")
xlim(0,2.5)
legend()
print("Máximo Local:")
print(max(ajuste1(corte1))) #Primer máximo que tiende a 1
datosExperimentales.append(max(ajuste1(corte1)))
```

## Segundo Corte

```python
ajuste2 = interp1d(corte22,corte2, kind = 'cubic') #Interpolación cubica.
```

```python
plot(corte22, ajuste2(corte22), color = "Red", label ="Interpolación\n Cúbica")
scatter(corte22,corte2, label = "Datos")
legend()
print("Mínimo Local:")
print(min(ajuste2(corte22))) #Primer mínimo que tiende a 0.
datosExperimentales.append(min(ajuste2(corte22)))
```

## Tercer Corte

```python
ajuste3 = interp1d(corte33,corte3, kind = 'cubic') #Interpolación cúbica.
plot(corte33, ajuste3(corte33), color = "Red", label ="Interpolacion\n Cúbica")
scatter(corte33,corte3, label="Datos")
legend()
print("Máximo Local:")
print(max(ajuste3(corte33))) #Máximo que tiende a 0.0175
datosExperimentales.append(max(ajuste3(corte33)))
```

## Cuarto Corte

```python
ajuste4 = interp1d(corte44,corte4, kind = 'cubic') #Interpolació cúbica.
plot(corte44, ajuste4(corte44), color = "Red", label ="Interpolación\n Cúbica")
scatter(corte44,corte4, label = "Datos")
legend()
print("Mínimo Local:")
print(min(ajuste4(corte44))) #Mínimo que tiende a 0.
print("Máximo Local:")
print(max(ajuste4(corte44))) #Máximo que tiende a 0.0042
datosExperimentales.append(min(ajuste4(corte44)))
datosExperimentales.append(max(ajuste4(corte44)))
```

## Quinto Corte

```python
ajuste5 = interp1d(corte55,corte5, kind = 'cubic') #Interpolación cúbica
plot(corte55, ajuste5(corte55), color = "Red", label = "Interpolación Cúbica")
scatter(corte55,corte5, label = "Datos")
legend(bbox_to_anchor=(1,0.4))
print("Mínimo Local:")
print(min(ajuste5(corte55))) #Mínimo que tiende a 0.
print("Máximo Local:")
print(max(ajuste5(corte55))) #Máximo que tiende a 0.0016
datosExperimentales.append(min(ajuste5(corte55)))
datosExperimentales.append(max(ajuste5(corte55)))
```

Por último se calculan los errores porcentuales entre los máximos y mínimos hallados por medio de interpolación y los datos teóricos.

```python
#Cálculo de error
errores = []
diferencia = []
for i in range(len(datosTeoricos)):
    teo = datosTeoricos[i]
    exp = datosExperimentales[i]
    error = (abs((exp - teo)/exp))*100
    errores.append(error)
    diferencia.append(abs(exp - teo))
```

A continuación se presentan los errores obtenidos al comparar los datos teóricos y experimentales. Cabe notar que fue necesario realizar dos comparaciones distintas: para los datos teóricos que no son cero, es decir los máximos, se toma como referencia el error porcentual experimental, mienras que para los mínimos, se toma la diferencia entre el valor teórico y el valor experimental. Esto se hace ya que el error porcentual es de 100$\%$ cuando el valor teórico es 0 y el experimental no. Ahora se presentan entonces estos valores de referencia en una matriz, organizada por columnas como: valor teórico, valor experimental, error porcentual y diferencia teórico-experimental.

```python
matrizResultados = zeros((7,4)) #Creación y llenado de la matriz.
for i in range(7):
    matrizResultados[i][0] = datosTeoricos[i]
    matrizResultados[i][1] = datosExperimentales[i]
    matrizResultados[i][2] = errores[i]
    matrizResultados[i][3] = diferencia[i]
```

```python
print("    Valor teórico   Valor Experimental    Error%         Diferencia")
print(matrizResultados) #Presentación de Resultados
```

Como es posible observar de la tabla anterior, los valores hallados por medio de la interpolación son bastante similares a los teóricos dados por el enucniado. Por un lado, los máximos presentan errores porcentuales iguales o menores al 31%, lo que se puede considerar un rango aceptable. Por otro lado, para los diferentes mínimos, las diferencias entre teórico y experimental son del orden de magnitud 0.00001 y menores, por lo que se podría afirmar igualmente que se encuentran en un rango de cercanía aceptable, los experimentales con respecto a los teóricos. Por último, más allá del análisis numérico, la comparación a simple vista de las gráficas obtenidas en esta tarea con respecto a las presentadas en el enunciado muestra que son bastante similares en cuanto a comportamiento y las posiciones de los picos y valles. Este tipo de comparación resulta más sencilla y efectiva cuando se usan los patrones de difracción también obtenidos. Y en estos las distintas simetrías presentes en los del enunciado se ven de igual forma en los obtenidos en la tarea.
