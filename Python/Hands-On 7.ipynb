```python
%pylab inline
from scipy.optimize import curve_fit
```

# Punto 1

```python
datos = genfromtxt("https://raw.githubusercontent.com/ComputoCienciasUniandes/MetodosComputacionalesDatos/master/hands_on/polyunkown.csv", delimiter = ",")
```

```python
abcisas=datos[:,0]
ordenadas=datos[:,1]
```

```python
def ajuste(x, a, b, c, d, e, f):
    return a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f
fitpars, covmat=curve_fit(ajuste, abcisas, ordenadas)
```

```python
scatter(abcisas, ordenadas)
plot(abcisas, ajuste(abcisas, *fitpars), color = "Red")
```

El grado del polinomio de mejor ajuste es 5.

## Punto 2

```python
#Creación de Arreglos
x=[2.3, 2.8, 3.2, 3.7, 4.3]
y=[34745, 19689, 12594, 7982, 5822]
```

```python
u = linspace(2,5,100) #Linspace sobre el cual graficar.
def ajuste(t, m): #Definir la función modelo.
    return (((0.2)*m)/((t)**3))
fitpars, covmat=curve_fit(ajuste, x, y) #Realizar ajuste.
```

```python
scatter(x, y, label = "Datos") #Plotear datos y ajuste.
plot(u, ajuste(u, *fitpars), color = "Red", label = "Ajuste Teórico", ls = "--")
xlim(2.2, 4.5)
legend()
print(*fitpars)
```

```python
#Intrapolación
r = linspace(2.3, 4.3, 100)
r2 = []
for i in range(len(r)):
    actual = (423658.8)/(r[i])**3
    r2.append(actual)
```

```python
scatter(r,r2, color = "Green", label = "Datos Intrapolados")
plot(u, ajuste(u, *fitpars), color = "Red", label = "Ajuste Teórico", ls = "--")
xlim(2.2, 4.5)
legend()
```
