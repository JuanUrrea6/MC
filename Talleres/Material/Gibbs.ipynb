Para comenzar el experimento, se crea una función de onda cuadrada, de alto $\frac{\pi}{4}$, que será utilizada como referencia para el ajuste de Fourier.

```python
#Creando una función escalonada 
x = linspace(0, 2*pi, 10000) #Linspace de X
y = []
for i in range(len(x)): #Creación por trozos de la función de onda cuadrada.
    if(x[i] <= pi):
        y.append(pi/4)
    if(x[i] > pi):
        if(x[i]<=2*pi):
            y.append(-pi/4)
```

```python
plot(x,y) #Se plotea para mejor visualización.
ylim(-1, 1)
title("Función de Onda Cuadrada")
xlabel("X")
ylabel("Y")
xlim(0, 2*pi)
```

Para continuar, se construye la función de seno cardinal iterativo, de manera que dependiendo del valor de entrada del parámetro t define la frecuencia.

```python
#Crear la función a ajustar
def sinC(x, t):
    return (sin((2*t + 1)*x))/(2*t + 1)
```

```python
#Función Iterativa
def ajustar(x,I):
    resultado = 0
    for i in range(I): #Suma del Anterior
        resultado = resultado + sinC(x,i)
    return resultado
```

```python
plot(x, ajustar(x,10), label ="10 Iteraciones")
plot(x, ajustar(x,5), label = "5 Iteraciones")
plot(x, ajustar(x,3), label = "3 Iteraciones")
plot(x, ajustar(x,2), label = "2 Iteraciones")
plot(x,y, color = "Black", label = "Onda Cuadrada")
title("Distintas Representaciones de Fourier")
xlabel("X")
ylabel("Y")
legend()
xlim(0,2*pi)
```

En la gráfica anterior se pueden observar las diferentes representaciones de Fourier para adaptarse a la función de onda cuadrada. Puede verse que, a mayor número de iteraciones, más similar es a dicha onda. Ahora, para determinar la relación existente entre el número de iteraciones y la diferencia entre el máximo y mínimo, se construyen funciones que permitan realizar las representaciones de Fourier en términos del número de iteraciones.

```python
def crearArregloSin(x, I): #Obtener el arreglo de valores en Y dado un arreglo e iteraciones.
    arreglo = [] 
    for i in range(len(x)):
        actual = ajustar(x[i], I) #Obtener cada término.
        arreglo.append(actual) #Añadir el termino al arreglo.
    return arreglo
```

```python
iteraciones = [] #Arreglo de Iteraciones
constantes = [] #Arreglo de Diferencias entre max y min.
for i in range(1,30):
    actual = crearArregloSin(x,i) #Obtener arreglo con la función anterior.
    maximo = max(actual) #Obtener máximos y mínimos.
    minimo = min(actual)
    diferencia = maximo - minimo #Calcular diferencia.
    iteraciones.append(i) #Actualizar los arreglos.
    constantes.append(diferencia)
```

```python
Gibbs=[]
x1 = linspace(0, 50,100)
for i in range(len(x1)):
    Gibbs.append(1.8519)

scatter(iteraciones,constantes, label = "Diferencia") #Graficar iteraciones y diferencias.
plot(x1, Gibbs, color = "Red", label = "Constante de Gibbs")
title("Iteraciones Vs Diferencia") #estética de la gráfica
xlabel("Iteraciones Realizadas")
ylabel("Diferencia entre Max y Min")
xlim(0, 30)
legend()
```

Como puede observarse en la gráfica anterior, el valor de la diferencia entre máximo y mínimo tiende a ser menor, y este decrecimiento se estabiliza hacia el valor de 1.85, el cual es aproximadamente la constante de Gibbs(1.8519).
