```python
%pylab inline
import smopy
from scipy.interpolate import *
import math
```

```python
#Importar los datos limpiados.
datos = genfromtxt("Coordenadas.txt", delimiter = ",")
```

Al realizar la limpieza del archivo de coordenadas, se dividieron con comas los valores de longitud, latitud, altura, hora, minuto y segundo, de manera que fuera más fácil su manejo por medio de arreglos separados.

```python
lon = datos[:,0] #Separar las columnas en arrays distintos.
lat = datos[:,1]
alt = datos[:,2]
hora = datos[:,3]
minuto = datos[:,4]
seg = datos[:,5]
```

```python
#Convertir tiempos a segundos
seg1 = 3600.*hora
seg2 = 60.*minuto
#Calcular tiempo total gastado
tiempo = []
for i in range(len(seg)):
    tiempo.append(seg[i] + seg1[i] + seg2[i])
```

```python
#Convertir a coordenadas cartesianas
theta = lon
phi = []
for i in range(len(lat)):
    actual = 90. - lat[i]
    phi.append(actual)
```

```python
SenPhi = [] #Calcular todos los senos y cosenos necesarios.
for i in range(len(phi)):
    actual = radians(phi[i])
    senoActual = sin(actual)
    SenPhi.append(senoActual)
```

```python
SenTheta = []
for i in range(len(theta)):
    actual = radians(theta[i])
    senoActual = sin(actual)
    SenTheta.append(senoActual)
```

```python
CosTheta = []
for i in range(len(theta)):
    actual = radians(theta[i])
    cosActual = cos(actual)
    CosTheta.append(cosActual)
```

```python
CosPhi = []
for i in range(len(phi)):
    actual = radians(phi[i])
    cosActual = cos(actual)
    CosPhi.append(cosActual)
```

```python
#Coordenadas Cartesianas.
X = []
Y = []
Z = []
for i in range(len(phi)):
    X.append(6371*SenPhi[i]*CosTheta[i])
    Y.append(6371*SenPhi[i]*SenTheta[i])
    Z.append(6371*CosPhi[i])
```

```python
#Normalizar espacio y tiempo
inicialX = X[0]
inicialY = Y[0]
inicialZ = Z[0]
inicialT = tiempo[0]
for i in range(len(X)):
    X[i] = X[i] - inicialX
    Y[i] = Y[i] - inicialY
    Z[i] = Z[i] - inicialZ
    tiempo[i] = tiempo[i] - inicialT
```

```python
#Producto Punto para proyectar
resultadoNorte = []
resultadoOriente = []
#Productos puntos sucesivos para proyectar sobre ambas direcciones.
for i in range(len(X)):
    orienteActual = sum(asarray([X[i],Y[i],Z[i]])*asarray([-sin(radians(-74.0758)), cos(radians(-74.0758)),0]))
    norteActual = sum(asarray([X[i], Y[i], Z[i]])*asarray([-cos(radians(-74.0758))*cos(radians(85.4019)), -sin(radians(-74.0758))*cos(radians(85.4019)), sin(radians(85.4019))]))
    resultadoNorte.append(norteActual)
    resultadoOriente.append(orienteActual)
```

```python
#Obtener mapa en las coordenadas.
mymap = smopy.Map((4.635, -74.089, 4.654, -74.070), z=15) #Importar mapa.
for i in range(len(resultadoNorte)): #Convertir coordenada a pixel.
    resultadoOriente[i], resultadoNorte[i] = mymap.to_pixels(lat[i], lon[i])
ax = mymap.show_mpl(figsize=(8, 6)) #Crear la figura.
ax.plot(resultadoOriente, resultadoNorte,'-k' , lw = 2, color = "Blue", label = "Recorrido") #Plotear el recorrido
ax.plot(resultadoOriente[-1], resultadoNorte[-1], 'or', ms = 10, mew=2, color = "Black", label = "Final") #Plotear punto final.
ax.plot(resultadoOriente[0], resultadoNorte[0], 'or', ms = 10, mew=2, label = "Inicio") #Plotear punto inicial.
title("Recorrido por el Parque")
legend(bbox_to_anchor=(1.4,1))
#savefig("Recorrido.png")
```

```python
#Calcular los diferenciales en X y Y.
difX = []
difY = []
difX.append(X[0]) #Incluir los ceros iniciales para el for.
difY.append(Y[0])
i = 1
while i < len(X):
    actualX = (X[i]) - (X[i-1])
    actualY = (Y[i]) - (Y[i-1])
    difX.append(actualX)
    difY.append(actualY)
    i+=1
```

```python
#Cálculo de Distancia Recorrida
distancia = []
i = 0
actual = 0
while i < len(X) :
    actual = actual + ((difX[i])**2 + (difY[i])**2)**(0.5)
    distancia.append(actual)
    i += 1
```

```python
horas = [] #Convertir tiempo a horas.
for i in range(len(tiempo)):
    actual = tiempo[i]/3600
    horas.append(actual)
```

```python
plot(horas,distancia) #Graficar para visualizar.
title("A walk in the park")
xlabel("t/horas")
ylabel("d/km")
xlim(0,1.6)
```

Aparentemente hay un error de apriximadamente 4 km en la distancia total recorrida. En caso de que no se pueda solucionar, queda aislado, pues en el mapa se pudo observar que el recorrido coincide con el que se buscaba obtener por el enunciado del taller. Sin embargo, hay que tomar en cuenta que la obtención del tiempo requerido para recorrer una distancia dada y las coordenadas en dicho instante se calculan a partir de este análisis de distancia.

```python
inerpolacion = interp1d(distancia, horas) #Interpolación distancia-horas
interLat = interp1d(horas, lat) #Interpolación a latitudes
interLon = interp1d(horas,lon) #Interpolación a Longitudes
```

Para las estimaciones de tiempo y coordenadas dada una distancia recorrida, se hace uso de interpolaciones entre cada set de datos involucrado.

```python
#Se obtienen los tiempos para cada distancia.
print(inerpolacion(1))
print(inerpolacion(2))
print(inerpolacion(3))
print(inerpolacion(4))
print(inerpolacion(5))
print(inerpolacion(6))
```

Para el cálculo de la velocidad, se hace uso de interpolación entre horas y distancia recorrida, de manera que se obtenga un arreglo de posición que se puede relacionar con el arreglo de horas dado por un linspace.

```python
#Cálculo de Distancia Recorrida
distancia = []
i = 0
actual = 0
while i < len(X) :
    actual = actual + ((difX[i])**2 + (difY[i])**2)**(0.5)
    distancia.append(actual)
    i += 1

horas = [] #Convertir tiempo a horas.
for i in range(len(tiempo)):
    actual = tiempo[i]/3600
    horas.append(actual)    
```

Para hallar entonces las rapideces, se utiliza el método de **forward difference** como se muestra en el notebook de la clase de Juan David. 

```python
#Hallar velocidad a partir de datos y a partir de interpolación.
func = interp1d(horas, distancia)
z = linspace(0, 1.6, len(distancia))
distancias = func(z)
h = 1.6/(len(distancia) - 1)
```

```python
distanciaF = np.roll(distancia,-1) #Realizar las diferencias como se mostró en clase.
distanciasF = np.roll(distancias,-1)
dD = (distanciaF - distancia)/h
dDD = (distanciasF - distancias)/h
dD = dD[:-1]
dDD = dDD[:-1]
z=z[:-1]
horas = horas[:-1]
```

```python
figure(figsize=(10,5))
ax2 = plt.subplot2grid((1,2), (0, 0), colspan=1)
plot(horas, dD)
ylabel("Rapidez(km/hora)")
xlabel("t/hora")
title("Rapideces basadas en Distancias")
xlim(0,1.6)
ax3 = plt.subplot2grid((1,2), (0, 1), colspan=1)
plot(z, dDD)
xlabel("t/horas")
ylabel("Rapidez(km/hora)")
title("Rapideces Basadas en Interpolación")
```

Como se había advertido en el enunciado, el resultado visualizado en la gráfica anterior resulta bastante irregular, ya que no es muy sencillo identificar "a ojo" las raideces en los distintos instantes de tiempo. Para intentar solucionar este problema, se busca enotnces hallar rapideces "promedio", es decir, haciendo el **h** más grande, de manera que ya no sería una derivada "infinitesimal", sino velocidades promedio entre puntos más alejados. Las dos gráficas anteriores se produjeron haciendo **forward difference**, para una usando el arreglo de distancias y el de horas, y para la otra usando la interpolación realizada a las distancias y un linspace para representar las horas.

```python
#Hallar velocidad mejorada
#Primero, función posición
func2 = interp1d(horas, distancia) #Definir la función de posición
e = linspace(0,1.6,100) #Linspace para recorrer. Ahora con menos valores que antes.
distancias2 = func2(e) #Arreglo de distancias.
h2 = 1.6/99 #Variación
```

```python
distancias2F = np.roll(distancias2,-1)
dD2 = (distancias2F-distancias2)/h2
dD2 = dD2[:-1]
e = e[:-1]
```

```python
plot(e, dD2)
xlim(0,1.6)
ylabel("Rapidez(km/hora)")
xlabel("t/horas")
```

En esta nueva gráfica resulta más sencillo visualizar las distintas rapideces a lo largo del tiempo de recorrido. Aunque es menos precisa que la anterior, permite ver el comportamiento promedio de las rapideces de forma más clara que antes.
